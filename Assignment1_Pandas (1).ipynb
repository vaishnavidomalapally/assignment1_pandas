## Assignment 2: Data Filtering & Transformation

- Display rides where CATEGORY == 'Business'.

- Show top 5 rides with the longest distance (MILES).

- Replace all missing PURPOSE values with "Not Specified".

- Create a new column TRIP_DURATION using END_DATE - START_DATE.

- Sort trips by distance in descending order.

```python
#to print all the rides where CATEGORY=BUSINESS
uber.loc[uber['CATEGORY']=="Business"] 
```

```python
#to print top 5 rides with longest distance(MILES)
a= uber.sort_values(by='MILES', ascending=False)
print(a.head(5))
```

```python
# replace all the null values in purpose column with not specified
uber['PURPOSE']=uber['PURPOSE'].fillna('Not Specified')
uber
```

```python
#to create a new column named TRIP_DURATION with END_DATE and START_DATE
uber['START_DATE'] = pd.to_datetime(uber['START_DATE'], errors='coerce')
uber['END_DATE'] = pd.to_datetime(uber['END_DATE'], errors='coerce')
uber['TRIP_DURATION'] = uber['END_DATE'] - uber['START_DATE']
uber
```

```python
#to sort trips by distance in descending order 
a= uber.sort_values(by='MILES', ascending=False)
a
```

## Assignment 3: Grouping and Aggregation

- Group by CATEGORY and find average miles per category.

- Find total trips for each PURPOSE.

- Identify top 3 start locations by number of rides.

- Plot:

- A bar chart of average miles by category.
- A pie chart of trip purposes.

```python
#Group by CATEGORY and find average miles per category
avg_miles = uber.groupby('CATEGORY')['MILES'].mean()
print(avg_miles)
```

```python
#Find total trips for each PURPOSE
total_trips = uber['PURPOSE'].value_counts()
print(total_trips)
```

```python
#Bar chart of average miles by category
import matplotlib.pyplot as plt

avg_miles.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Average Miles by Category')
plt.ylabel('Average Miles')
plt.xlabel('Category')
plt.xticks(rotation=45)
plt.show()
```

```python
#Pie chart of trip purposes
total_trips.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
plt.title('Trip Purposes')
plt.ylabel('')  # hide y-label
plt.show()
```
